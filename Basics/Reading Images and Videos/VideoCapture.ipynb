```python
import cv2
```

```python
capture=cv2.VideoCapture(0)
```

```python
while True:
    has_frame,frame=capture.read()
    print(has_frame, frame)
    if not has_frame:
        print('can\'t get frame')
        break
    cv2.imshow('frame', frame)
    key=cv2.waitKey(3)
    if key==27:
        print('pressed esc')
        break

capture.release()
cv2.destroyAllWindows()
```
